# Matrix Plots

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
tips.corr()
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fc=flights.corr()
```

```python
sns.heatmap(fc,annot=True,cmap='coolwarm')
```

```python
dp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(dp,annot=False,cmap='coolwarm')
```

```python
sns.heatmap(dp,cmap='magma')
```

```python
sns.heatmap(dp,cmap='magma',linecolor='white',linewidths=1)
```

```python
sns.heatmap(dp,cmap='magma',linecolor='black',linewidths=1)
```

```python
sns.clustermap(dp)
```
